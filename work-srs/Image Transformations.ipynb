# 1. Transform

```python
# Takes PIL image or numpy array
# Creates a frame around input array and rotates by specified angle
# Crops resulting array and outputs it
def rotandcrop(PILimg, angle):
    '''
    Input PILimg: PIL image or numpy array
    angle: angle by which to rotate image
    Output: rotated and cropped numpy array
    '''
    
    #if type(PILimg) is PIL.JpegImagePlugin.JpegImageFile: # for jpegs
    #    img = np.mean(PILimg, -1) # convert to 2D array and colourshift
    #else:
    #    img = PILimg;
    
    img = PILimg
    size = img.shape
    maxsize = max(size)
    pad = int(2* maxsize)
    
    avg = np.mean(img)    

    buffered = np.zeros((pad, pad)) # create buffer array larger than image
    buffered[:] = avg # Choose which value to surround img by
    buffered[ int(pad/2 - size[0]/2):int(pad/2 + size[0]/2), int(pad/2 - size[1]/2):int(pad/2 + size[1]/2) ] = img

    rotated = Image.fromarray(buffered).rotate(angle) # convert to image and rotate using PIL
    rsize = rotated.size
    cropsize = (rsize[1]/7, rsize[1]/7, 6*rsize[0]/7, 6*rsize[0]/7) # arbitrary. Find better way of defining crop
    cropped = rotated.crop(cropsize)
    result = np.array(cropped)
    
    #rotated = scipy.ndimage.rotate(img, angle) # automatically crops, does not allow for background change
    
    return result
```

```python
# Create new xVec and yVec.  To get the dimensions right we need to know the rotation angle. 

# get the number of pixels in the transformed image
numYpts, numXpts = h5transformed.shape

# get the range of the original data in metres
yMax = max(yVec)
xMax = max(xVec)

# calculate the range of the rotated image
rotationAngle = 45 # AT THE MOMENT THE ROTATION ALWAYS INCREASES THE SIZE TO MAXIMUM. IF IMAGE IS CROPPED FIRST THIS SHOULD BE DELETED
rotAngleDeg = np.deg2rad(rotationAngle)
yMaxNew = xMax * np.sin(rotAngleDeg) + yMax * np.cos(rotAngleDeg)
xMaxNew = xMax * np.cos(rotAngleDeg) + yMax * np.sin(rotAngleDeg)

# create the new xVec and yVec
yVecNew = np.linspace(0., yMaxNew, numYpts, dtype=float)
xVecNew = np.linspace(0., xMaxNew, numXpts, dtype=float)
```
